```python
ep = var('ε')
lam = var('λ')
c = var('c')
Q = var('Q')
x,y = var('x y')
val = var('b')
show(x)
ep=-1
val=3/2
```

```python
Func = (val)*(2-ep)*x^3 + Q*np.sqrt((val))*ep^2*x^2 +(3/2)*(1-2*ep)*x - np.sqrt(val)*Q*ep
Func2 = (val)*(2*ep-1)*x^3 + Q*np.sqrt(val)*x^2 +(val)*(ep-2)*x/ep - np.sqrt(val)*Q/ep
show(Func2)
```

```python
show(solve(Func2==0,x))
```

```python
xprime = -3*x/ep + np.sqrt(val)*lam*y^2/ep - np.sqrt(val)*Q/ep*(1-ep*x^2-y^2)+val*x*(1-ep*x^2-y^2)+3*ep*x^3
```
