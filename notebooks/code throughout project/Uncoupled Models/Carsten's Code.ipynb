```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# initial redshift
z0 = 10**14

# Hubble today (in units of 100 km/sMpc)
h = 0.7

# critical density in GeV^4

rho_cr = 8.09687*h*h*10**(-47)
show(rho_cr)

# density parameter for matter and radiation today

Omega_matter_0 = 0.3
Omega_radiation_0 = 2.47*10**(-5)/(h*h)

# densities of matter and radiation today in GeV^4

rho_matter0 = Omega_matter_0*rho_cr
rho_radiation0 = Omega_radiation_0*rho_cr

# 1 GeV in (reduced) M_Planck (which is set to 1!)

GeV = 1/(2.435*10**18)

# getdensities in M_Planck units
rho_matter0 = rho_matter0*GeV**4
show(rho_matter0)
rho_radiation0 = rho_radiation0*GeV**4

#print(rho_matter0,rho_radiation0)

# Rough approximation of initial densities at z0 - ignores that rho_m does not scale like a^{-3}!
rho_mi = rho_matter0*(z0**3)
rho_ri = rho_radiation0*(z0**4)

# parameter of the theory: 

# Potential parameter: double exponential potential V = V_1*exp(-lambda1*phi) + V_2*exp(-lambda2*phi)
lambda1 =  1.0
lambda2 =  0.0
V_1 = 0.01*10**(-120)
V_2 = -0.1*10**(-126)

# Coupling: Parameter in coupling function 
alpha = 0

# initial conditions for field
phi_ini = .5
phi_prime_ini = 0

phi_i = [phi_ini, phi_prime_ini]


# Potential
def V(phi):

    V = V_1*np.exp(-lambda1*phi) + V_2*np.exp(-lambda2*phi)

    return V

# Derivative of Potential
def Vp(phi):
    
    Vp = -lambda1*V_1*np.exp(-lambda1*phi) - lambda2*V_2*np.exp(-lambda2*phi)

    return Vp

# Conformal factor C(phi)

def C(phi):
    
    # for now: (1 + exp(2*alpha*phi))/2
    # return (1 + np.exp(2*alpha*phi))/2
    return np.exp(2*np.cosh(alpha*phi))

# Derivative of conformal factor dC/dphi

def Cp(phi):
    # return alpha*np.exp(2*alpha*phi)
    return 2*alpha*np.sinh(alpha*phi)*C(phi)

# Coupling C'/C
def beta(phi):
    return Cp(phi)/(2*C(phi))
    
# Equations of motions

def cosmo(u,N):
  
    phi, phi_prime = u # u[0] = phi, u[1] = phi_prime
    
    rho_mat = rho_mi*C(phi)/C(phi_ini)*(np.exp(-3*N))
    rho_rad = rho_ri*(np.exp(-4*N))
    
    H2 = (V(phi) + rho_mat + rho_rad)/(3 - 0.5*phi_prime**2)
    HdotoH2 = -0.5*(phi_prime**2 + ( (4*rho_rad)/3 + rho_mat )/H2 )
    
    u0prime = phi_prime 
    u1prime = - (3 + HdotoH2)*phi_prime - Vp(phi)/H2 - beta(phi)*rho_mat/H2
    
    return[u0prime, u1prime]
    

N_final = 35

N = np.linspace(0,N_final,10000)

sol = odeint(cosmo, phi_i, N)

phi_sol = sol[:,0]
phi_prime_sol = sol[:,1]

Pot = V_1*np.exp(-lambda1*phi_sol)

coup = alpha + alpha*np.tanh(alpha*phi_sol)

rho_m = rho_mi*np.exp(-3*N)*C(phi_sol)/C(phi_ini)
rho_r = rho_ri*np.exp(-4*N)

H2 = (Pot + rho_m + rho_r)/(3 - (phi_prime_sol**2)/2)
#show(H2)
H0 = np.sqrt(H2)/1.75040e-63 # from Richard to convert into km/s Mpc
#show(H0)

Omega_matter = rho_m/(3*H2)
Omega_radiation = rho_r/(3*H2)
Omega_phi = (phi_prime_sol**2)/6 + Pot/(3*H2)
show(Omega_phi)
# get the redshift
z = (np.exp(N_final)/np.exp(N)) - 1

# # Plot the results

fig, axs = plt.subplots(3)
# plt.text(1,1e12,r"$\\alpha=${alpha}  $\\lambda_1=${lambda1}     $\phi_i=${phi_ini}", fontsize=15)
plt.xlabel(r'$N$')

axs[0].plot(N, phi_sol, 'b', label=r'$\phi$')
axs[0].set_ylabel(r'$\phi$')


axs[1].plot(N, coup, label=r'$\beta$')
axs[1].set_ylabel(r'$\beta$')
axs[1].set_ylim([0,2.5*alpha])


# axs[1].axhline(y=0.05, c='r', ls='--') #point at which needs to be reached 0.05


axs[2].plot(N, Omega_radiation, label= r'$\Omega_{\gamma}$')
axs[2].plot(N, Omega_matter, label=r'$\Omega_{m}$' )
axs[2].plot(N, Omega_phi, label=r'$\Omega_{\phi}$')

axs[2].set_ylim([0.01,1.5])

axs[2].legend()
# plt.show()
```
