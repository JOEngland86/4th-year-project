# Plotting Evolution of Cosmological Parameters

# 1. Plotting evolution of Klein Gordon Equation:  
Klein Gordon equation of a scalar potential field, $\phi$ is represented by: $ \ddot{\phi} + 3H\dot{\phi} + \frac{dV}{d\phi}$ where $\dot{}$ represents the time derivative. In this case, we want to differentiate $\phi$ w.r.t $N = ln(a)$ which is denoted by a '. $\phi^{'} = \frac{d\phi}{dN}$ is related to $\dot{\phi}$ by the hubble parameter such as $$\dot{\phi} = H\phi^{'} $$

```python
import numpy as np
import sympy as sym
```

```python
# 1. Define Potential V

# a) Define variables used in V:
phi = var(r'φ') # scalar field
lam = var(r'λ') # 
k = var(r'κ') # k = sqrt(8πG)
V0 = var(r'V_0')

# b) Define V function and its derivative
V = function('V')(phi)
V = V0*exp(-lam*k*phi)
dVdφ  = V.diff(phi)
```

```python
# 2. Change φ in dVdΦ to φ(Ν)
N = var('N')
Phi = function(r'Φ')(N)
dvdΦ = dVdφ.subs(phi == Phi)
```

# Change of variables of Klein Gordon Equation:
If we re-write the klein gordon equation in terms of N, we get:
$$ \phi'' + \phi'\left[3 + \frac{2\dot{H}}{H^2}\right] + \frac{1}{H^2}\frac{dV}{d\phi} $$

$\dot{H}$ is equal to $$-\frac{\kappa^2}{2}\left(\gamma\rho_{c,0}\Omega_{\gamma,0} a^{-4} + (\dot{\phi})^2 \right)$$ is the time derivative of the Hubble parameter. We can we-write this w.r.t N by replacing $\dot{\phi}$ with $(H \phi')$.

We can then re-write $\frac{2\dot{H}}{H^2}$ as $-(\kappa \phi')^2 - 3\frac{\gamma (H_0)^2 \Omega_{\gamma,0}}{a^4 H^2}$. We can substitute this into the Klein Gordon Equation, before we substitute the expression for $H^{-2}$. We must also define all the new variables these equations bring into the 2nd order differential equation. 

```python
#3. Write the Klein Gordon Equation w.r.t N
# a) Define key variables and constants in Klein Gordon equation
H = var('H') # Hubble parameter
a = var('a') # scale factor
gam = var(r'γ')
H0 = 67.7 # Hubble constant in units 
ome_m = var(r'Ω_m')
ome_r = var(r'Ω_r')
HdotH2 = var('HdotH2')

# a) Enter Klein Gordon Equation w.r.t to N:
KG = Phi.diff(N,2) + (3+HdotH2)*Phi.diff(N,1) + H^(-2)*dvdΦ

# b) Enter expression for 1/H^2 and 2Hdot/H^2
inverse_H_squared = (2/k - (Phi.diff(N))^2)/(V.substitute(phi ==Phi) + (3*H0^2/k^2)*a^(-3)*(ome_m+ome_r/a))
y = -(k*Phi.diff(N,1))^2 - 3*H^(-2)*(gam*H0^2*ome_r)/(a^4)
```

```python
# 3. Substitute expression for Hubble Parameter H as also depends on Φ:

# c) Substitute expression into Klein Gordon Equation:
KG1 = KG.substitute(HdotH2==y, H^(-2) ==inverse_H_squared)
show(KG1)
```

```python
# 4. Substitute Constants and solve Klein Gordon equation

# a) Constant values
constants = [lam == 0.5,
             V0 == 4, 
             ome_m == 0.31, 
             ome_r == 0, 
             k ==sqrt(8*np.pi), # Assumed using cosmological units so G = 1
             a == e^N
            ] 

show(KG1.substitute(ome_r==0))
# c) substitute constants and solve Klein Gordon Equation
KG2 = KG1.substitute(constants)
#show(KG2)
```
