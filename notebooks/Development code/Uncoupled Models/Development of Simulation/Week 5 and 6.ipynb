```python
# Klein Gordon Equation:
v = function('v')(phi)
H = function('H')(phi)
KG = diff(phi,N,2) + 2*diff(phi,N,1) + (1/H^2)*diff(v,phi,1)
show(KG)
```

```python
def klein_Gordon(Q,H,v,x):
    QN = function('{}'.format(Q))(x)
    d2Qdx2 = QN.diff(x,2)
    dQdX = 2*QN.diff(x,1)
    Qsym = var('{}'.format(Q)) # At this step have to change Q from a function of N to a variable to be able to differentiate V
    dvdQ = (1/H^2)*v.diff(Qsym)
    return d2Qdx2 + dQdX + dvdQ
```

```python
phi = var('φ')
H = var('H')
General_KG = klein_Gordon(phi,H,v,N)
```
